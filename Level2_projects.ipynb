```python
expenses = []

def add_expense(expenses, amount, category, date, description):
    expense = {
        'amount': amount,
        'category': category,
        'date': date,
        'description': description
    }
    expenses.append(expense)
```

Usage:

```python
add_expense(expenses, 50.0, 'Groceries', '2024-01-25', 'Bought fruits and vegetables')
add_expense(expenses, 20.0, 'Transport', '2024-01-24', 'Taxi fare')
```

**Week 2:**

```python
def read_and_add_expense(expenses):
  amount = float(input("Enter the amount: "))
  category = input("Enter the category: ")
  date = input("Enter the date (YYYY-MM-DD): ")
  description = input("Enter a description: ")

  if (amount == None or category == None or date == None or description == None):
    return "One of the values was not entered"

  expense = {
      'amount': amount,
      'category': category,
      'date': date,
      'description': description
        }
  expenses.append(expense)

  print("Expense added successfully")

def display_expenses(expenses):
    for expense in expenses:
        print(f"Date: {expense['date']}, Category: {expense['category']}, "
              f"Amount: ${expense['amount']:.2f}, Description: {expense['description']}")

```

Usage:

```python
read_and_add_expense(expenses)

display_expenses(expenses)
```

**Week 3:**

```python
def filter_by_date(expenses, target_date):
    filtered_expenses = []
    for expense in expenses:
        if expense['date'] == target_date:
            filtered_expenses.append(expense)
    return filtered_expenses

def filter_by_category(expenses, target_category):
    filtered_expenses = []
    for expense in expenses:
        if expense['category'].lower() == target_category.lower():
            filtered_expenses.append(expense)
    return filtered_expenses

def filter_by_amount(expenses, min_amount, max_amount):
    filtered_expenses = []
    for expense in expenses:
        if min_amount <= expense['amount'] <= max_amount:
            filtered_expenses.append(expense)
    return filtered_expenses
```

Usage:

```python
filtered_by_date = filter_by_date(expenses, '2024-01-25')
filtered_by_category = filter_by_category(expenses, 'Transport')
filtered_by_amount = filter_by_amount(expenses, 10, 30)

print(filtered_by_date)
print(filtered_by_category)
print(filtered_by_amount)
```

**Week 4:**

```python
def calculate_total_expenses(expenses):
    total = sum(expense['amount'] for expense in expenses)
    return total

def calculate_average_spending(expenses):
    if not expenses:
        return 0
    average = sum(expense['amount'] for expense in expenses) / len(expenses)
    return average

def category_with_highest_expense(expenses):
    category_totals = {}
    for expense in expenses:
        category = expense['category']
        category_totals[category] = category_totals.get(category, 0) + expense['amount']

    if not category_totals:
        return None

    highest_expense_category = max(category_totals, key=category_totals.get)
    return highest_expense_category
```

Usage:

```python
total = calculate_total_expenses(expenses)
average = calculate_average_spending(expenses)
highest_expense_category = category_with_highest_expense(expenses)

print(total)
print(average)
print(highest_expense_category)
```

**Week 5:**

```python
def set_saving_goal(target_amount, target_date=None):
    saving_goal = {
        'target_amount': target_amount,
        'target_date': target_date
    }
    return saving_goal

def allocate_earnings(earnings, saving_goal, saving_percentage):
    saving_allocation = earnings * (saving_percentage / 100)
    remaining_for_expenses = earnings - saving_allocation

    # Check if the saving allocation exceeds the target amount
    if saving_allocation + (saving_goal.get('current_amount', 0)) > saving_goal['target_amount']:
        saving_allocation = saving_goal['target_amount'] - saving_goal.get('current_amount', 0)
        remaining_for_expenses = earnings - saving_allocation

    allocation = {
        'saving_allocation': saving_allocation,
        'remaining_for_expenses': remaining_for_expenses
    }
    return allocation
```

Usage:

```python
my_saving_goal = set_saving_goal(5000, '2024-12-31')

total_earnings = 3000
saving_percentage = 20  # Allocate 20% of earnings to saving
allocation = allocate_earnings(total_earnings, my_saving_goal, saving_percentage)

print(my_saving_goal)
print(allocation)
```

```python
def join_team():
    team = input("Choose your team (A/B): ").upper()
    while team not in ['A', 'B']:
        print("Invalid team. Please choose 'A' or 'B'.")
        team = input("Choose your team (A/B): ").upper()
    return "Team A" if team == 'A' else "Team B"

def get_current_team_turn(turn_count):
    return "Team A" if turn_count % 2 == 0 else "Team B"

team_scores = {"Team A": 0, "Team B": 0}

def validate_answer(user_answer, correct_answer, current_team, points_for_correct=10, points_for_incorrect=0):
    if user_answer.lower().strip() == correct_answer.lower().strip():
        team_scores[current_team] += points_for_correct
        return points_for_correct, "Correct!"
    else:
        team_scores[current_team] += points_for_incorrect
        return points_for_incorrect, "Incorrect. The correct answer was: " + correct_answer
```

# Project 4


**Week 1**:


```python
cookbook = []

def add_recipe(recipes, name, ingredients, steps, cuisine, prep_time):
    recipe = {
        'name': name,
        'ingredients': ingredients,
        'steps': steps,
        'cuisine': cuisine,
        'prep_time': prep_time
    }
    cookbook.append(recipe)
```

**Week 2**:

```python
def create_and_add_recipe(recipes):
    name = input("Enter the recipe name: ")
    num_ingredients = int(input("How many ingredients? "))
    ingredients = []
    for _ in range(num_ingredients):
        item = input("Enter ingredient name: ")
        quantity = input("Enter ingredient quantity: ")
        ingredients.append({"name": item, "quantity": quantity})

    num_steps = int(input("How many preparation steps? "))
    steps = []
    for _ in range(num_steps):
        step = input("Enter preparation step:")
        steps.append(step)

    cuisine = input("Enter the type of cuisine: ")
    prep_time = input("Enter the preparation time: ")

    add_recipe(recipes, name, ingredients, steps, cuisine, prep_time)
    print(f"Recipe for {name} added successfully.")

def search_recipe(recipes, search_name):
    search_name = search_name.strip().lower()
    for recipe in recipes:
        if recipe['name'].strip().lower() == search_name:
            return recipe
    failure = f"Recipe '{search_name}' not found. Please check the spelling and try again."
    return failure

```

Usage:

```python
create_and_add_recipe(cookbook)
print(search_recipe(cookbook, "Cupcake"))
print(search_recipe(cookbook, "Hot dogs"))
```

**Week 3**:

```python
def display_all_recipes(recipes):
    if not recipes:
        print("No recipes found.")
        return
    for recipe in recipes:
        print(f"Name: {recipe['name']}")
        print(f"Cuisine: {recipe['cuisine']}")
        print(f"Prep Time: {recipe['prep_time']}")
        print("Ingredients:")
        for ingredient in recipe['ingredients']:
            print(f" - {ingredient['quantity']} of {ingredient['name']} )")
        print("Steps:")
        for i, step in enumerate(recipe['steps'], start=1):
            print(f" {i}. {step}")
        print("-" * 40)  # Separator for readability

def edit_recipe_attribute(recipes, recipe_name, attribute, new_value):
    for recipe in recipes:
        if recipe['name'].lower() == recipe_name.lower():
            if attribute in recipe:
                recipe[attribute] = new_value
                print(f"{recipe_name}'s {attribute} has been updated.")
                return
            else:
                print(f"Attribute '{attribute}' not found in the recipe.")
                return
    print(f"Recipe '{recipe_name}' not found.")

def delete_recipe(recipes, recipe_name):
    for i, recipe in enumerate(recipes):
        if recipe['name'].lower() == recipe_name.lower():
            del recipes[i]
            print(f"Recipe '{recipe_name}' has been deleted.")
            return
    print(f"Recipe '{recipe_name}' not found.")
```

Usage:

```python
display_all_recipes(cookbook)
edit_recipe_attribute(cookbook, 'Cupcake', 'prep_time', '45 minutes')
delete_recipe(cookbook, 'Cupcake')
```

**Week 4**:

```python
def display_recipes_by_prep_time(recipes, min_prep_time, max_prep_time):
    found = False
    for recipe in recipes:
        # Extract numeric part from prep_time and convert to integer
        prep_time_minutes = int(recipe['prep_time'].replace('min', ''))
        if min_prep_time <= prep_time_minutes <= max_prep_time:
            print(f"Name: {recipe['name']}, Prep Time: {recipe['prep_time']}, Cuisine: {recipe['cuisine']}")
            found = True
    if not found:
        print(f"No recipes found within {min_prep_time} to {max_prep_time} minutes prep time.")

def display_recipes_by_cuisine(recipes, cuisine_type):
    found = False
    for recipe in recipes:
        if recipe['cuisine'].lower() == cuisine_type.lower():
            print(f"Name: {recipe['name']}, Prep Time: {recipe['prep_time']}")
            found = True
    if not found:
        print(f"No recipes found for {cuisine_type} cuisine.")
```

Usage:

```python
display_recipes_by_prep_time(cookbook, 15, 30)
display_recipes_by_prep_time(cookbook, 45, 50)
display_recipes_by_cuisine(cookbook, 'American')
display_recipes_by_cuisine(cookbook, 'Italian')
```

**Week 5**:

```python
def get_user_ingredients():
    ingredients = []
    print("Enter the ingredients you have (type 'done' when finished):")

    while True:
        ingredient = input().strip().lower()
        if ingredient == 'done':
            break
        ingredients.append(ingredient)

    return ingredients

def display_recipes_by_ingredients(recipes, user_ingredients):
    found = False
    for recipe in recipes:
        recipe_ingredients = {ingredient['name'].lower() for ingredient in recipe['ingredients']}
        if recipe_ingredients.issubset(set(user_ingredients)):
            print(f"Name: {recipe['name']}, Prep Time: {recipe['prep_time']}, Cuisine: {recipe['cuisine']}")
            found = True

    if not found:
        print("No recipes found matching your available ingredients.")
```

Usage;

```python
user_ingredients = get_user_ingredients()
display_recipes_by_ingredients(cookbook, user_ingredients)
```
