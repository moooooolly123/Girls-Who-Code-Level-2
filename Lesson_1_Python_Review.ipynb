#Loops

In Python, we have two types of loops:

*   while loops: run while a certain condition is true
*   for loops: run through every part of a string

Let's start with our while loop.

Print the variable *text* three times. Remember that we gave text a value earlier.

Hint: The variable *times* keeps track of how many times we printed text. Add one to *times* each time you print text

```python
times = 0

while times != 3:
  print(v)
  times += 1
```
